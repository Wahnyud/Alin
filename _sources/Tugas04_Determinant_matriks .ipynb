```python
def determinant_2x2(matrix):
    a11 = matrix[0][0]
    a12 = matrix[0][1]
    a21 = matrix[1][0]
    a22 = matrix[1][1]

    det_A = a11 * a22 - a12 * a21
    return det_A

matrix_A = [
    [1, 2],
    [3, 4]
]

determinant = determinant_2x2(matrix_A)
print("Determinan dari matriks A:", determinant)
```

##Menghitung det mat dan invers mat dan Menghitung invers matriks 5x5 dengan kofaktor yang di transposekan

Untuk menghitung invers matriks
5×5 menggunakan metode kofaktor yang ditransposekan, ikuti langkah-langkah berikut:

Langkah 1:

Hitung Determinan Matriks
Untuk menemukan invers suatu matriks, Anda pertama-tama harus memastikan bahwa determinannya tidak sama dengan nol. Determinan matriks
𝐴
A dapat dihitung menggunakan metode ekspansi kofaktor. Misalkan
𝐴
A adalah matriks
5
×
5:

\begin{pmatrix}
a_{11} & a_{12} & a_{13} & a_{14} & a_{15} \\
a_{21} & a_{22} & a_{23} & a_{24} & a_{25} \\
a_{31} & a_{32} & a_{33} & a_{34} & a_{35} \\
a_{41} & a_{42} & a_{43} & a_{44} & a_{45} \\
a_{51} & a_{52} & a_{53} & a_{54} & a_{55} \\
\end{pmatrix} \]

Determinannya, \( \det(A) \), dihitung dengan ekspansi kofaktor pada baris pertama (atau baris/kolom lain):
\[ \det(A) = \sum_{j=1}^{5} (-1)^{1+j} a_{1j} \det(A_{1j}) \]
Di mana \( A_{1j} \) adalah submatriks \( 4 \times 4 \) yang diperoleh dengan menghapus baris pertama dan kolom \( j \) dari matriks \( A \).
### Langkah 2: Hitung Matriks Kofaktor
Untuk setiap elemen \( a_{ij} \) dari matriks \( A \), hitung kofaktornya, \( C_{ij} \), yang didefinisikan sebagai:
\[ C_{ij} = (-1)^{i+j} \det(A_{ij}) \]
Di mana \( A_{ij} \) adalah submatriks \( 4 \times 4 \) yang diperoleh dengan menghapus baris \( i \) dan kolom \( j \) dari \( A \).
### Langkah 3: Bentuk Matriks Kofaktor
Buat matriks kofaktor \( C \) dari semua kofaktor \( C_{ij} \):
\[ C = \begin{pmatrix}
C_{11} & C_{12} & C_{13} & C_{14} & C_{15} \\
C_{21} & C_{22} & C_{23} & C_{24} & C_{25} \\
C_{31} & C_{32} & C_{33} & C_{34} & C_{35} \\
C_{41} & C_{42} & C_{43} & C_{44} & C_{45} \\
C_{51} & C_{52} & C_{53} & C_{54} & C_{55}
\end{pmatrix} \]
### Langkah 4: Transpos Matriks Kofaktor
Transposkan matriks kofaktor untuk mendapatkan matriks adjugat \( \text{adj}(A) \):
\[ \text{adj}(A) = C^T \]
### Langkah 5: Hitung Invers Matriks
Akhirnya, invers dari matriks \( A \) diberikan oleh:
\[ A^{-1} = \frac{1}{\det(A)} \text{adj}(A) \]
### Contoh
Misalkan \( A \) adalah:
\[ A = \begin{pmatrix}
2 & -1 & 0 & 0 & 1 \\
1 & 3 & 2 & 0 & -1 \\
0 & 1 & -1 & 4 & 2 \\
3 & 0 & -2 & 1 & 3 \\
1 & 2 & 3 & 1 & -1
\end{pmatrix} \]

Langkah ini untuk menghitung determinan, matriks kofaktor, dan matriks adjugat secara manual sangat panjang, sehingga disarankan menggunakan alat bantu seperti software matematika (contohnya MATLAB, Python dengan NumPy, atau kalkulator matriks) untuk mengurangi risiko kesalahan dan mempercepat proses perhitungan.

```python
import numpy as np

# Definisikan matriks 5x5
A = np.array([
    [2, -1, 0, 0, 1],
    [1, 3, 2, 0, -1],
    [0, 1, -1, 4, 2],
    [3, 0, -2, 1, 3],
    [1, 2, 3, 1, -1]
])

# Pastikan determinan tidak nol
det_A = np.linalg.det(A)
if det_A == 0:
    raise ValueError("Matriks tidak memiliki invers karena determinannya nol.")

# Hitung invers matriks menggunakan fungsi bawaan NumPy
A_inv = np.linalg.inv(A)

print("Matriks A:")
print(A)
print("\nInvers dari matriks A:")
print(A_inv)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menggambar vektor
def plot_vector(v, color='b'):
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=color)

# Vektor asli
v = np.array([2, 3])

# 1. Rotasi
def rotation_matrix(theta):
    return np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])

theta = np.pi / 4  # Rotasi 45 derajat
R = rotation_matrix(theta)
v_rotated = R.dot(v)

# 2. Skala
def scaling_matrix(sx, sy):
    return np.array([
        [sx, 0],
        [0, sy]
    ])

sx, sy = 2, 3  # Skala x dengan 2 dan y dengan 3
S = scaling_matrix(sx, sy)
v_scaled = S.dot(v)

# 3. Geser
def shear_matrix(k, axis='x'):
    if axis == 'x':
        return np.array([
            [1, k],
            [0, 1]
        ])
    elif axis == 'y':
        return np.array([
            [1, 0],
            [k, 1]
        ])

k = 1  # Faktor geser
Sh = shear_matrix(k, axis='x')
v_sheared = Sh.dot(v)

# 4. Refleksi
def reflection_matrix(axis='x'):
    if axis == 'x':
        return np.array([
            [1, 0],
            [0, -1]
        ])
    elif axis == 'y':
        return np.array([
            [-1, 0],
            [0, 1]
        ])
    elif axis == 'xy':
        return np.array([
            [0, 1],
            [1, 0]
        ])

axis = 'x'
Ref = reflection_matrix(axis)
v_reflected = Ref.dot(v)

# Plot semua vektor
plt.figure()
plot_vector(v, 'b')
plot_vector(v_rotated, 'r')
plot_vector(v_scaled, 'g')
plot_vector(v_sheared, 'm')
plot_vector(v_reflected, 'c')

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend(['Original', 'Rotated', 'Scaled', 'Sheared', 'Reflected'])
plt.show()
```
