* 1.Penjelasan tentang Matriks:

  Matriks adalah tabel berisi elemen-elemen yang diatur dalam baris dan kolom. Setiap elemen dalam matriks dapat diidentifikasi oleh posisi baris dan kolomnya. Misalnya, matriks A dengan ordo m x n memiliki m baris dan n kolom.

  Secara umum
  Matriks digunakan secara luas dalam berbagai bidang, termasuk matematika, ilmu komputer, fisika, ekonomi, dan banyak lagi. Matriks sangat berguna untuk merepresentasikan data dalam bentuk yang terstruktur dan dapat diterapkan dalam berbagai operasi matematika seperti penjumlahan, pengurangan, perkalian, dan operasi lainnya

```python
import numpy as np
A = np.array([[1, 3, 3],[2, 1, 9],[2, 4, 7]])
B = np.array([[1,6,3],[2,8,7],[3,9,5]]) #3x3
C = np.array([[2,4],[5,2],[6,7]]) #3x2
D = np.array([[3,7,5],[1,4,8],[2,4,5]])
E = A+B #penjumlahan
F = 2*A #Skala
G = A@C #perkalian
H = B@C
I = A@H
J = A@B
K = J@C
L = A-B #pengurangan
```

```python
C_trans = C.transpose()
Z = C_trans@C
print(E)
print('\n')
print(F)
print('\n')
print(G)
print('\n')
print(I)
print('\n')
print(K)
print('\n')
print(L)
print('\n')
print(C_trans)
print(Z)
```

```python
# Membuat matriks A dengan ordo 3x3
A = [[4,8,1],
     [9,0,2],
     [2,5,-3]]

# Membuat matriks B dengan ordo 3x3
B = [[3,8,2],
     [5,7,-2],
     [6,3,7]]

# Membuat matriks C dengan ordo 3x2
C = [[0, 0, 0],
     [0, 0, 0],
     [0, 0, 0]]

# Membuat matriks D dengan ordo 3x2
D = [[0, 0, 0],
     [0, 0, 0],
     [0, 0, 0]]


# Mengalikan matriks A dan B
for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range(len(B)):
            C[i][j] += A[i][k] * B[k][j]

# Menampilkan hasil perkalian
print("Hasil perkalian matriks A dan B:")
for row in C:
    print(row)
```
