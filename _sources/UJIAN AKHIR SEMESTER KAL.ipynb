## 1.Hitunglah PageRank dari graph berikut

Picture1.png

```python
import numpy as np

# Buat matriks ketetanggaan
nodes = ['A', 'B', 'C', 'D', 'E']
adj_matrix = np.array([
    [0, 1, 1, 0, 0],  # A
    [1, 0, 0, 1, 1],  # B
    [1, 0, 0, 1, 1],  # C
    [0, 1, 1, 0, 1],  # D
    [0, 1, 1, 1, 0],  # E
])

# Hitung PageRank
def pagerank(matrix, alpha=0.85, max_iter=100, tol=1e-6):
    n = matrix.shape[0]
    M = matrix / np.sum(matrix, axis=0)
    M = np.nan_to_num(M)
    v = np.random.rand(n, 1)
    v = v / np.linalg.norm(v, 1)
    e = np.ones((n, 1))
    for _ in range(max_iter):
        v_new = alpha * np.dot(M, v) + (1 - alpha) / n * e
        if np.linalg.norm(v_new - v, 1) < tol:
            break
        v = v_new
    return v.flatten()

# Panggil fungsi PageRank
pr_values = pagerank(adj_matrix)

# Buat dictionary hasil
pr_dict = {nodes[i]: pr_values[i] for i in range(len(nodes))}

# Cetak hasil
print(pr_dict)
```

### JAWABAN

1. Inisialisasi Matriks

* Matriks Augmented (A): Matriks augmented A berisi koefisien dan konstanta dari sistem persamaan linear.

```python
A = np.array([[1, -1, 1, 3],
              [2, 1, 3, 8],
              [6, 0, -1, 4]], dtype=float)
```

* Matriks Salinan (E): Membuat salinan dari matriks A untuk dilakukan operasi.

```python
E = np.copy(A)
```

2. Reduksi ke Bentuk Eselon Baris

*   Loop Utama (i): Loop ini berjalan untuk setiap kolom i dari matriks.



```python
for i in range(3):
```

* Menemukan Baris Pivot: Dalam setiap iterasi, kita mencari baris dengan nilai absolut terbesar di kolom i untuk dijadikan pivot.

```python
baris_pivot = i
for j in range(i+1, 3):
    if np.abs(E[j, i]) > np.abs(E[baris_pivot, i]):
        baris_pivot = j
```

* Menukar Baris: Menukar baris pivot dengan baris saat ini i.

```python
E[[i, baris_pivot]] = E[[baris_pivot, i]]
```

* Normalisasi Baris Pivot: Membagi seluruh elemen di baris pivot dengan elemen pivot sehingga elemen pivot menjadi 1.

```python
E[i] = E[i] / E[i, i]
```

* Eliminasi Baris di Bawah: Mengeliminasi semua elemen di bawah pivot dalam kolom i dengan melakukan operasi baris.

```python
for j in range(i+1, 3):
    E[j] = E[j] - E[j, i] * E[i]
```

3. Reduksi ke Bentuk Eselon Baris Tereduksi
* Loop Utama (i): Loop ini berjalan mundur dari baris terakhir ke baris pertama.

```python
for i in range(2, -1, -1):
```

* Eliminasi Baris di Atas: Mengeliminasi semua elemen di atas pivot dalam kolom i dengan melakukan operasi baris.

```python
for j in range(i-1, -1, -1):
    E[j] = E[j] - E[j, i] * E[i]
```

4. Menampilkan Solusi
* Matriks Solusi (X): Membuat salinan dari matriks E yang sudah tereduksi.

```python
X = np.copy(E)
```

* Menampilkan Bentuk Eselon Baris Tereduksi:

```python
print("Bentuk eselon baris tereduksi dari matriks augmented adalah:")
print(X)
```

* Ekstrak dan Menampilkan Solusi: Menampilkan solusi dari sistem persamaan linear.

```python
solusi = X[:, -1]
print("Solusi: ", solusi)
```

*   BERIKUT CODE LENGKAPNYA


```python
import numpy as np

# Matriks augmented
A = np.array([[1, -1, 1, 3],
              [2, 1, 3, 8],
              [6, 0, -1, 4]], dtype=float)

# Matriks untuk menyimpan hasil
E = np.copy(A)

# Loop untuk mereduksi matriks E ke bentuk eselon baris
for i in range(3):
    # Temukan baris pivot
    baris_pivot = i
    for j in range(i+1, 3):
        if np.abs(E[j, i]) > np.abs(E[baris_pivot, i]):
            baris_pivot = j
    # Tukar baris pivot dengan baris saat ini
    E[[i, baris_pivot]] = E[[baris_pivot, i]]

    # Normalisasi baris pivot
    E[i] = E[i] / E[i, i]

    # Eliminasi baris di bawah
    for j in range(i+1, 3):
        E[j] = E[j] - E[j, i] * E[i]

# Loop untuk mereduksi matriks E ke bentuk eselon baris tereduksi
for i in range(2, -1, -1):
    for j in range(i-1, -1, -1):
        E[j] = E[j] - E[j, i] * E[i]

# Matriks solusi
X = np.copy(E)

print("Bentuk eselon baris tereduksi dari matriks augmented adalah:")
print(X)

# Ekstrak solusi
solusi = X[:, -1]
print("Solusi: ", solusi)
```
